# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [2000, 2005, 2010, 2015, 2020, 2025]
y = [2, 4, 6, 8, 10, 12]
```

## LinePlot

```python
#LinePlot
plt.plot(x, y)
plt.show()
```

## Scatter Plot

```python
# Scatter Plot
plt.scatter(x, y)
plt.show()
```

## Histogram

```python
plt.hist(x, bins=12)
plt.show()
```
